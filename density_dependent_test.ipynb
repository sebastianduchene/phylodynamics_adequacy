# Test for diversity-dependent speciation

```r
library(TreePar)
library(TreeSim)
library(NELSI)
```

```r
set.seed(1)
```

```r
time <- c(0,1)
rho <- c(0.5,1)

tree<-sim.rateshift.taxa(100,1, lambda = c(2,0.1), mu = c(0,0.05),frac=c(1,1),times=time,complete=FALSE)
plot(tree[[1]])
```

```r
x <- sort(getx(tree[[1]]),decreasing=TRUE)
```

```r
resDD <- bd.densdep.optim(x, discrete = FALSE, continuous = T)[[2]]
```

```r
resDD
```

```r
resShifts <- bd.shifts.optim(x,c(rho,1),0.1,0.1,2.1)[[2]][[2]]
```

```r
resDD
```

```r
resShifts
```

```r
# Best model where AIC smallest
AICDD <- 2*3+2*resDD$value
```

```r
AICShifts <- 2*5+2*resShifts[1]
```
