```r
library(NELSI)
library(TreeSim)
library(TreePar)
```

```r
#Simulate with the following parameters
R0 <- 2
death <- 0.5
rho <- 0.5

lambda <- R0*death

tr <- sim.bdsky.stt(n = 100, lambdasky = lambda, deathsky = death, timesky = 0, sampprobsky = rho, model = 'SIR', 
                   N = 1000, trackinfecteds = T)
tr_2 <- addroot(tr[[1]], tr[[1]]$root.edge)
```

```r
par(mfrow = c(1, 2))
plot(tr[[1]], show.tip.label = F)
#nodelabels(round(intnode.times(tr[[1]]), 2), cex = 0.5)
#tiplabels(round(allnode.times(tr[[1]], tipsonly = T), 2), cex = 0.5)

plot(tr[[2]]$eventtimes, tr[[2]]$infecteds, xlim = rev(range(tr[[2]]$eventtimes)), pch = 20, col = rgb(1, 0, 0, 0.5),
    ylim = c(min(tr[[2]]$recovereds, tr[[2]]$infecteds, tr[[2]]$susceptibles), 
             max(tr[[2]]$recovereds, tr[[2]]$infecteds, tr[[2]]$susceptibles)))
points(tr[[2]]$eventtimes, tr[[2]]$susceptibles, pch = 20, col= rgb(0, 0, 0, 0.3))
points(tr[[2]]$eventtimes, tr[[2]]$recovereds, pch = 20, col= rgb(0, 1, 0, 0.5))
points(tr[[2]]$eventtimes, tr[[2]]$cumulativesampleds, pch = 20, col= rgb(0, 0, 1, 0.5))
```

```r
x <- getx(tr_2, sersampling = T)
out <- bdsky.stt.optim(x = x[, 1], ttype = x[, 2], sampprob = c(0.25, 1), constdeath = 0, root = 0)
```
