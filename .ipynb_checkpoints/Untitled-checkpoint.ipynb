```r
likdif <- vector()
estimates <- c()
for (index in 1:5){

    tip_type <- x[, 2]
    test_tree <- trees[[index]]    
    droptip<-c("EBOV|KM034550|EM095|SierraLeone_EM|2014-05-25","EBOV|KM034554|G3676|SierraLeone_G|2014-05-27","EBOV|KM034559|G3680|SierraLeone_G|2014-05-28","EBOV|KM034563|G3687|SierraLeone_G|2014-05-28","EBOV|KM034561|G3683|SierraLeone_G|2014-05-28","EBOV|KM034562|G3686|SierraLeone_G|2014-05-28")
    test_tree <- drop.tip(test_tree, droptip)

    x <- getx(test_tree, sersampling = 1)
    rootheight <- max(x[, 1])
    
    if(numb_re == 1){
        out <- optim(c(2, 1, 0.7), LikShiftsSTTebolaConst, times = x, ttype = tip_type, 
             sprobc = sprobc, deathfix = death_fix, cutoff = 0)
    }else{
        out <- optim(c(rep(2, numb_re), 1, 0.7), LikShiftsSTTebola, times = x, ttype = tip_type,
                sprobc = sprobc, deathfix = death_fix, cutoff = 0)
    }
    
    estimates <- rbind(estimates, parepi(out, sprobc = sprobc, deathfix = death_fix))
}
estimatesraw <- estimates
```
